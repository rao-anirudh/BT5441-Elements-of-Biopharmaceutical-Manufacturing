```python
rates_df = pd.read_csv("/content/BT5441_Assignment2_RateData.csv")
```

```python
rates_df.columns = ["Cell line"] + list(rates_df.columns[1:])
```

```python
rates_df
```

```python
X = rates_df.drop("Cell line",axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
from sklearn.decomposition import PCA
pca = PCA()
X_pca = pca.fit_transform(X)
```

```python
pca_df = pd.DataFrame(X_pca)
pca_df.columns = ["PC"+str(i+1) for i in range(len(pca_df.columns))]
pca_df["Cell line"] = rates_df["Cell line"]
```

```python
pca_df
```

```python
import seaborn as sns
```

```python
plt.figure(dpi=150)
ax = sns.barplot(x=range(1,5),y=100*pca.explained_variance_ratio_)
ax.bar_label(ax.containers[0], fontsize=10);
plt.ylim(0,75)
plt.xticks(labels=range(1,5), ticks=range(0,4))
plt.ylabel("Percentage of explained variance")
plt.xlabel("Principal component");
```

```python
plt.figure(dpi=150, figsize=(2.5,2.5))
sns.scatterplot(x=pca_df["PC1"],y=pca_df["PC2"],hue=pca_df["Cell line"])
plt.axhline(y=0, color='black', linestyle='-', linewidth=0.5)
plt.axvline(x=0, color='black', linestyle='-', linewidth=0.5)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend(bbox_to_anchor=(1.2, 1))
plt.show()
```

```python
plt.figure(dpi=150, figsize=(5,0.5))
sns.scatterplot(x=abs(pca_df["PC1"]),y=0,hue=pca_df["Cell line"])
plt.ylim(-0.1,0.1)
plt.xlim(0,6)
plt.yticks([])
plt.xlabel("|PC1|")
plt.legend(bbox_to_anchor=(1.05, 1.1))
plt.show()
```

```python
plt.figure(dpi=150, figsize=(5,0.5))
sns.scatterplot(x=abs(pca_df["PC2"]),y=0,hue=pca_df["Cell line"])
plt.axvline(x=0, color='black', linestyle='-', linewidth=0.5)
plt.xlim(0,6)
plt.ylim(-0.1,0.1)
plt.yticks([])
plt.xlabel("|PC2|")
plt.legend(bbox_to_anchor=(1.3, 1.1))
plt.show()
```

```python
explained_variances = pca.explained_variance_ratio_
data_loadings = pca.components_.T * np.sqrt(explained_variances)
data_pc_vector = pd.DataFrame(data_loadings)
data_pc_vector.index = rates_df.drop("Cell line",axis=1).columns
data_pc_vector.columns = [f'PC {i+1} Loading' for i in range(0,4)]
data_pc_vector
```

```python
plt.figure(dpi=150, figsize=(9,9))
sns.scatterplot(x=pca_df["PC1"],y=pca_df["PC2"],hue=pca_df["Cell line"])
plt.axhline(y=0, color='black', linestyle='-', linewidth=0.5)
plt.axvline(x=0, color='black', linestyle='-', linewidth=0.5)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend(bbox_to_anchor=(1.15, 1))

for i, feature in enumerate(rates_df.drop("Cell line",axis=1).columns):

    if feature == "Specific growth rate":
        feature = "Specific\ngrowth rate"
    elif feature == "Specific productivity":
        feature = "Specific\nproductivity"

    plt.annotate(feature,
                 (data_loadings[i, 0]*10,
                  data_loadings[i, 1]*10),
                 color='black', xytext=(2, -2),textcoords='offset points')

    plt.arrow(0, 0,
              data_loadings[i, 0]*10,
              data_loadings[i, 1]*10,
              color='black',
              alpha=0.25,
              width=0.007,
             )

plt.show()
```
